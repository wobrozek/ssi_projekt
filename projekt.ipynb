```python
class ProcessingData:
    @staticmethod
    def splitSet(x,k): #k - l. zmiennoprzecinkowa np. k=0.7
        n = int(len(x)*k) #numer przecicia wzgldem wartoci procentowej (k=0.7 -> 70%)
        xTrain = x[:n] #podzia na wartoci treningowe i testowe
        xVal = x[n:] 
        return xTrain, xVal
    
    def splitSetDwa(x,k): #k - l. zmiennoprzecinkowa np. k=0.7
        n = int(len(x)*k) #numer przecicia wzgldem wartoci procentowej (k=0.7 -> 70%)
        xTrain = x[:n] #podzia na wartoci treningowe i testowe
        return xTrain

diabetes=ProcessingData.splitSetDwa(diabetes,0.3)
diabetesT,diabetesV=ProcessingData.splitSet(diabetes,0.3)
```
